```python
import pandas as pd
import xlwings as xw
```

```python
df1= pd.read_excel('input1.xlsx')
#data1=df1.values.tolist()
```

```python
df2=pd.read_excel('input2.xlsx')
#data1=df2.values.tolist()
```

```python
df1['total statements']=df2['total_statements']
df1['total reasons']=df2['total_reasons']
df1
```

```python
df1['total']= df1['total statements']+df1['total reasons']
df1['name adj']=df1['Name'].str.lower()
df1
```

```python
df1.sort_values(by=['total','name adj'], ascending=[False,True],inplace=True)
```

```python
df1
```

```python
out1=df1.copy()
out1.drop('name adj',axis=1,inplace=True)
out1.drop('Team Name',axis=1,inplace=True)
out1.drop('User ID',axis=1,inplace=True)
out1.drop('total',axis=1,inplace=True)
out1.rename(columns={'total statements': 'No of Statements'}, inplace=True)
out1.rename(columns={'total reasons': 'No of reasons'}, inplace=True)
l=[]
for i in range(1,22):
    l.append(i)
l
out1['S No']=l
out1
out1.to_excel('output1.xlsx',index=False)
```

```python
df1['team adj']=df1['Team Name'].str.lower()
df1.sort_values(by='S No',ascending=True,inplace=True)
mymap={}
newdata=df1['team adj']
ori_data=df1['Team Name']

newdata=list(newdata)
ori_data=list(ori_data)
mymap={}
for i in range(len(newdata)):
    mymap[newdata[i]]=ori_data[i]

print(mymap)
```

```python
column1_name='total statements'
column2_name='total reasons'
df1_grouped = df1.groupby(df1['Team Name'].str.lower())[[column1_name, column2_name]].mean()
df1_grouped['total']=df1_grouped['total statements']+df1_grouped['total reasons']
df1_grouped.sort_values(by='total',ascending=False,inplace=True)
#df1_grouped['New Team Name']=df1_grouped['Team Name']
df1_grouped = df1_grouped.reset_index()     #Reset the index to convert the group labels into columns

df1_grouped['Team Name'] =df1_grouped['Team Name'].map(mymap)
df1_grouped.drop('total',axis=1,inplace=True)
df1_grouped.rename(columns={'total statements': 'Average Statements'}, inplace=True)
df1_grouped.rename(columns={'total reasons': 'Average reasons'}, inplace=True)
df1_grouped
```

```python
list=[]
for i in range(len(df1_grouped['Team Name'])):
    list.append(i+1)

df1_grouped.insert(0,'S No',list)
df1_grouped
```

```python
df1_grouped.to_excel('output2.xlsx',index=False)
```
